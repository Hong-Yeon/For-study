```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
frequency = [100, 200, 300, 400]
frequency = list(map(int,frequency))
frequency 
```

```python
a = []
t = np.linspace(0,1,500)
for i in range(len(frequency)):
     a.append(np.sin(2*t*np.pi*frequency[i]))   
plt.figure(figsize = (24,8))
plt.plot(t,a[i])
```
